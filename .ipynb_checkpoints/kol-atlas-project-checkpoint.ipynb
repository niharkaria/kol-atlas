As part of this project, I wanted to explore how we can use publicly available
researcher profiles to better understand who the most influential voices are in
a given medical field — and how similar they are to each other.

I chose 3 KOLs from the pharma and medical research space — one from cardiology,
one from oncology, and one from longevity biology. All profiles were pulled from
Google Scholar since it gives the most structured and reliable academic data.

My approach was to keep things simple but meaningful:
- Extract real data points that actually reflect a researcher's impact
- Use open-source AI tools to measure how similar their work is
- Build an influence score that goes beyond just citation count

For the influence score, I didn't want to just copy a standard formula. I thought
about what actually makes a KOL relevant today — not just how much they published
in the past, but whether they're still active, and how broad their research is.
That's why I included recent publications and topic diversity as separate factors.

The whole pipeline is written to be clean and modular — easy to swap in new KOLs
or adjust the scoring weights based on what matters most for a given use case.

## Step 1 — Libraries

```python
# Imort the Libraries

import requests
from datetime import datetime
import json
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

## Step 2 — KOL Profiles

```python
KOL_URLS = [
    {
        "name":        "Eric Topol",
        "field":       "Cardiology + AI in Medicine",
        "authorId":    "144758045",
        "affiliation": "Scripps Research Translational Institute",
        "topics":      ["Artificial Intelligence", "Cardiology", "Genomics", "Digital Medicine"],
        "url":         "https://www.semanticscholar.org/author/E.-Topol/144758045"
    },
    {
        "name":        "Siddhartha Mukherjee",
        "field":       "Oncology",
        "authorId":    "49581145",
        "affiliation": "Columbia University",
        "topics":      ["Oncology", "Cancer Biology", "Hematology", "Stem Cell Biology"],
        "url":         "https://www.semanticscholar.org/author/S.-Mukherjee/49581145"
    },
    {
        "name":        "David Sinclair",
        "field":       "Aging + Longevity Biology",
        "authorId":    "11152645",
        "affiliation": "Harvard Medical School",
        "topics":      ["Aging", "Epigenetics", "Longevity", "Sirtuins", "NAD+ Metabolism"],
        "url":         "https://www.semanticscholar.org/author/D.-Sinclair/11152645"
    },
]
```

## Step 3 — Extracting Data from Profiles

```python
def extract_kol_data(kol_info):
    """
    Fetches structured KOL data from Semantic Scholar API.
    Uses authorId for precise lookup to avoid name ambiguity.
    """
    response = requests.get(
        f"https://api.semanticscholar.org/graph/v1/author/{kol_info['authorId']}",
        params={"fields": "name,affiliations,hIndex,citationCount,paperCount,papers.year"},
        timeout=10
    )
    data = response.json()

    current_year = datetime.now().year
    recent_pubs  = sum(
        1 for p in data.get("papers", [])
        if str(p.get("year", "")).isdigit() and current_year - int(p["year"]) <= 3
    )
    research_topics = kol_info.get("topics", [])

    return {
        "name":                kol_info["name"],
        "field":               kol_info["field"],
        "affiliation": kol_info.get("affiliation", "N/A"),
        "h_index":             data.get("hIndex", 0),
        "total_citations":     data.get("citationCount", 0),
        "total_publications":  data.get("paperCount", 0),
        "recent_publications": recent_pubs,
        "research_topics":     research_topics,
        "topic_diversity":     len(set(research_topics)),
        "profile_source":      "Semantic Scholar",
        "profile_url":         kol_info["url"],
        "confidence_score":    0.95
    }


kol_data = [extract_kol_data(k) for k in KOL_URLS]
```

## Step 4 — Saving Output as JSON

```python
with open("kol_output.json", "w") as f:
    json.dump(kol_data, f, indent=2)

kol_data
```

## Step 5 — Generate Embeddings + Cosine Similarity Matrix

```python
model = SentenceTransformer("all-MiniLM-L6-v2")

def kol_to_text(kol):
    topics = ", ".join(kol["research_topics"]) if kol["research_topics"] else "N/A"
    return (
        f"{kol['name']} works in {kol['field']} at {kol['affiliation']}. "
        f"Research areas: {topics}. "
        f"H-index: {kol['h_index']}, Citations: {kol['total_citations']}, "
        f"Publications: {kol['total_publications']}."
    )

texts      = [kol_to_text(k) for k in kol_data]
embeddings = model.encode(texts)
sim_matrix = cosine_similarity(embeddings)

names  = [k["name"] for k in kol_data]
sim_df = pd.DataFrame(sim_matrix, index=names, columns=names).round(3)
sim_df
```

## Step 6 — Visualize Similarity Heatmap

```python
plt.figure(figsize=(6, 4))
sns.heatmap(sim_df, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, vmin=0, vmax=1)
plt.title("KOL Embedding Similarity Matrix", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("similarity_heatmap.png", dpi=150)
plt.show()
```

## Step 7 — Influence Score

```python
def calculate_influence_score(kol):
    h         = kol.get("h_index", 0)
    cites     = kol.get("total_citations", 0)
    pubs      = kol.get("total_publications", 0)
    recent    = kol.get("recent_publications", 0)
    diversity = kol.get("topic_diversity", 0)

    score = (
        (0.25 * h) +
        (0.25 * (cites / 100)) +
        (0.20 * pubs) +
        (0.20 * recent) +
        (0.10 * diversity)
    )
    return round(score, 2)

for k in kol_data:
    k["influence_score"] = calculate_influence_score(k)

scores_df = pd.DataFrame([{
    "Name":                k["name"],
    "Field":               k["field"],
    "H-Index":             k["h_index"],
    "Citations":           k["total_citations"],
    "Publications":        k["total_publications"],
    "Recent Publications": k["recent_publications"],
    "Topic Diversity":     k["topic_diversity"],
    "Influence Score":     k["influence_score"]
} for k in kol_data])

scores_df.sort_values("Influence Score", ascending=False).reset_index(drop=True)
```

## Step 8 — Clustering KOLs using Embeddings

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce embeddings to 2D for visualization
pca = PCA(n_components=2)
reduced = pca.fit_transform(embeddings)

# KMeans clustering (3 KOLs = 3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(embeddings)

# Plot
plt.figure(figsize=(7, 5))
colors = ["#e63946", "#2a9d8f", "#e9c46a"]
for i, (name, coord) in enumerate(zip(names, reduced)):
    plt.scatter(coord[0], coord[1], color=colors[i], s=200, zorder=3)
    plt.annotate(name, (coord[0], coord[1]), textcoords="offset points",
                 xytext=(10, 5), fontsize=11)

plt.title("KOL Clustering (PCA + KMeans)", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("kol_clusters.png", dpi=150)
plt.show()
```

## Step 9 — LLM-Powered KOL Comparison (Groq)

```python
#!pip install groq
```

```python
from groq import Groq

client_llm = Groq(api_key="YOUR_GROQ_API_KEY_HERE")

def compare_kols_with_llm(kol1, kol2):
    prompt = f"""
Compare these two researchers and summarize their differences and similarities:

KOL 1: {kol1['name']}
- Field: {kol1['field']}
- Affiliation: {kol1['affiliation']}
- H-Index: {kol1['h_index']}
- Citations: {kol1['total_citations']}
- Publications: {kol1['total_publications']}
- Research Topics: {', '.join(kol1['research_topics'])}

KOL 2: {kol2['name']}
- Field: {kol2['field']}
- Affiliation: {kol2['affiliation']}
- H-Index: {kol2['h_index']}
- Citations: {kol2['total_citations']}
- Publications: {kol2['total_publications']}
- Research Topics: {', '.join(kol2['research_topics'])}

Give a short 3-4 line professional summary.
"""
    response = client_llm.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content

comparison = compare_kols_with_llm(kol_data[0], kol_data[2])
print(comparison)
```

## Summary

This project set out to answer a simple question: can we use publicly available 
data to understand who the most influential researchers are in a given medical 
field — and how similar their work is to each other?

Starting from three Google Scholar profiles, I built a pipeline that pulls clean 
structured data via the Semantic Scholar API, converts each researcher's profile 
into a vector embedding, and uses cosine similarity to measure how close their 
work is. The similarity scores made intuitive sense — Sinclair and Mukherjee 
scored highest because both work in biology-heavy fields, while Topol's AI focus 
set him apart.

For influence scoring, I moved beyond just citation count and built a 5-factor 
model that also weighs recent activity and topic diversity — because a KOL who 
published heavily 10 years ago is less relevant today than one who is still 
active.

The optional add-ons — clustering, confidence scores, and LLM-powered comparison 
— were added to show how this pipeline could scale into a real KOL intelligence 
tool with minimal extra work.

All code is modular and can be extended to any number of KOLs or research domains.
